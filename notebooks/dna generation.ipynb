```python

import random

base_pairs = ['A','T', 'C','G']

def generate_fragments(N,L, scaling_factor = 10, print_only = False):
    fragments = []
    for n in range(N):
        new_fragment = ''.join(random.choices(base_pairs, weights=[0.15,0.15,0.35,0.35], k=L*scaling_factor))
        if print_only:
            print(new_fragment)
        else:
            fragments.append(new_fragment)
    return fragments
    

def verify_fragments(fragments):
    for f in fragments:
        cg_perc = (f.count('C') + f.count('G'))/len(f)
        print(cg_perc,f )

        if cg_perc < 0.69 or cg_perc > 0.71:
            return False
fragments = generate_fragments(5,1000)
verify_fragments(fragments)
fragments = generate_fragments(5,1000, scaling_factor = 1000000, print_only=False)
```
